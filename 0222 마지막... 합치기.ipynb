```python
import torch
import numpy as np
import pandas as pd
```

```python
csvs = [pd.read_csv(f'results2/0011-ECATF5/submission-0011-ECATF5-{i}.csv') for i in range(1, 7)]
```

```python
csvs[0]
```

```python
nps = [csv.to_numpy() for csv in csvs]
```

```python
nps[0].shape
```

```python
dps = np.zeros_like(nps[0])
```

```python
dps[:, 0] = nps[0][:, 0]
```

```python
for p in nps:
    dps[:, 1:] += p[:, 1:]
```

```python
dps[:, 1:] /= 6
```

```python
dps[:, 0] = dps[:, 0].astype(np.int)
```

```python
df = pd.DataFrame(dps, columns=csvs[0].columns)
df
```

```python
df.id = df.id.astype(np.int)
```

```python
df
```

```python
df.to_csv('results2/0011-ECATF5/submission-1,2,3,4,5,6.csv', index=False)
```
