```python
from scipy.ndimage.filters import gaussian_filter
```

```python
def replicate_padding(x):
    r = x[:, 300:]
    l = x[:, :300]
    return np.concatenate([r, x, l], axis=1)


def random_shift(x):
    shift = random.randint(0, 600)
    return np.roll(x, shift, axis=1)


def random_amplitude(x):
    amplitude = random.random() * 0.4 + 0.8
    return x * amplitude


def random_sin(x):
    freqs = [100, 120, 150, 200, 300]
    wave = np.sin(np.array(list(range(600))) / random.sample(freqs, 1)[0] * np.pi)
    amplitude = random.random() * 0.1
    signal = 1 + wave * amplitude
    return x * signal.reshape(1, -1)


def random_cos(x):
    freqs = [100, 120, 150, 200, 300]
    wave = np.cos(np.array(list(range(600))) / random.sample(freqs, 1)[0] * np.pi)
    amplitude = random.random() * 0.1
    signal = 1 + wave * amplitude
    return x * signal.reshape(1, -1)


def random_blur(x):
    sigma = random.random() * 0.3
    return gaussian_filter(x, sigma=sigma)


def omnirandom(x):
    x = random_shift(x)
    x = random_sin(x)
    x = random_cos(x)
    x = random_amplitude(x)
    x = replicate_padding(x)
    x = random_blur(x)
    return x
```

```python
z = omnirandom(x)
for i in [0, 2, 4, 1, 3, 5]:
    plt.figure(figsize=(15, 3))
    for j in range(3):
        plt.subplot(1, 3, j + 1)
        plt.plot(x[i * 3 + j], "g-")
        plt.plot(z[i * 3 + j], "b-")
        plt.title(titles[i] + "_" + suffix[j])
    plt.tight_layout()
    plt.show()
```
