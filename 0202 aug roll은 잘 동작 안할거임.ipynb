```python
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
data = np.load("data/0201.npz")
```

```python
X_train = data["X_train"]
```

```python
X_train.shape
```

```python
x = torch.tensor(X_train[250])
```

```python
for i in range(3):
    plt.figure(figsize=(24, 4))
    for j in range(6):
        plt.subplot(1, 6, j + 1)
        plt.plot(x[i * 6 + j])
    plt.tight_layout()
    plt.show()
```

```python
x_roll100 = torch.roll(x, 100, dims=1)
x_roll100[:, :100] = 0
```

```python
for i in range(3):
    plt.figure(figsize=(24, 4))
    for j in range(6):
        plt.subplot(1, 6, j + 1)
        plt.plot(x_roll100[i * 6 + j])
    plt.tight_layout()
    plt.show()
```
